```python
import pandas as pd
 
df = pd.read_csv('Gaana-Data-Analysis/songs.csv')
```

```python
df
```

##1. Checking Null Values


```python
df.isnull().sum()
```

##2.Data Preprocess

#####2.1 Converting durtion(mm:ss) into seconds

```python
duration_lst = []

for i in df['duration']:
   min,sec = i.split(':')[0],i.split(':')[1]
   s = (int(min)*60)+int(sec)

   duration_lst.append(s)

df['duration']=duration_lst
df.head()
```

##3.Data Analysis

3.1 Average Duration of a song throughout the dataset


```python
print("Average duration of a song is ",df['duration'].mean())
```

3.2 Total number of languages


```python
for i in set(df['language']):
  print(i)
len(set(df['language']))
```

 Number of hindi songs present in dataset


```python
c = 0 
for i in df['language']:
  if (i=='Hindi'):
    c+=1
print("No of Hindi songs", c)
```

```python
len(df['language']=='Hindi')
```

Number of songs  per language

```python
for i in df['language'].unique():
  print(i,len(df['language']==i))
```

Number of unique singer 

```python
all_singer= []
for i in df['singer']:
  all_singer += i.split('|')
print(len(set(all_singer)))
```

```python
data =[]

for i in set(all_singer):
  c=0
  for j in df['singer']:
    if (i in j):
      c+=1
  data.append([i,c])
```

```python
data
```

```python
#creating a new dataFrame of songs with singer and number of songs sang
```

```python
songs =  pd.DataFrame(data,columns = ['singer','songs'])
```
